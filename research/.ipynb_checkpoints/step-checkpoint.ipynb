```python
# https://github.com/chris-santiago/steps 
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
import statsmodels.datasets
from statsmodels.api import OLS
from statsmodels.tools import add_constant

from steps.forward import ForwardSelector
from steps.subset import SubsetSelector
```

```python
df = pd.read_csv('copy_us_owid_covid_monthly.csv')
pd.set_option('display.max_columns', None)
df = df.dropna()
df.head()
```

```python
x = df.iloc[:, 1:]
y = df['cci']
```

```python
selector = ForwardSelector(normalize=True, metric='aic')
selector.fit(x, y)
ForwardSelector(normalize=True)
x.loc[:, selector.best_support_]
```

```python
selector = ForwardSelector(normalize=True, metric='bic')
selector.fit(x, y)
ForwardSelector(normalize=True)
x.loc[:, selector.best_support_]
```

```python
selector = SubsetSelector(normalize=True, metric='aic')
selector.fit(x, y)
SubsetSelector(normalize=True)
x.loc[:, selector.get_support()]
```

```python
selector = SubsetSelector(normalize=True, metric='bic')
selector.fit(x,y)
SubsetSelector(normalize=True)
x.loc[:, selector.get_support()]
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.feature_selection import RFE
np.set_printoptions(suppress=True)
```

```python
y=df['cci']
x=df.drop(['cci'],axis=1)
model=LinearRegression()
sfs=SequentialFeatureSelector(model,direction='backward')
sfs.fit(x,y)
sfs.get_feature_names_out()
```

```python
rfe = RFE(model)             
rfe = rfe.fit(x, y)
list(zip(x.columns,rfe.support_,rfe.ranking_))
```

```python
rfe_columns = x.columns[rfe.support_]
rfe_columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
df = pd.read_csv('copy_us_owid_covid_monthly_cop.csv')

pd.set_option('display.max_columns', None)
df = df.dropna()
df.head()
```

```python
# Add a constant
x = [ 'positive_rate', 'stringency_index', 'total_cases', 'total_deaths','umsent','cci_lag1','VaxFull','new_cases','new_deaths','VaxFull^2',
       'stringency_index_stan_start_start']
x = df[x]
x = sm.add_constant(x)
y = df['cci']
       
# Build the model with RFE features
model = sm.OLS(y,x).fit()

#Summary of linear model
print(model.summary())
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Dropping variables
- High p-value/High VIF - drop the variable
- High p-value/Low VIF - drop the variable w/ high p-value first
- Low p-value/Low VIF - accept the variable

```python
# Add a constant
x = [ 'positive_rate_stan', 'total_deaths','umsent',
       'stringency_index']
x = df[x]
x = sm.add_constant(x)
y = df['cci']
       
# Build the model with RFE features
model = sm.OLS(y,x).fit(cov_type='HC3')

#Summary of linear model
print(model.summary())
```

```python
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.iolib.summary2 import summary_col
x = [1, 3, 5, 6, 8, 3, 4, 5, 1, 3, 5, 6, 8, 3, 4, 5, 0, 1, 0, 1, 1, 4, 5, 7]
y = [0, 1, 0, 1, 1, 4, 5, 7,0, 1, 0, 1, 1, 4, 5, 7,0, 1, 0, 1, 1, 4, 5, 7]
d = { "x": pd.Series(x), "y": pd.Series(y)}
df = pd.DataFrame(d)
df['xsqr'] = df['x']**2  
mod = smf.ols('y ~ x', data=df)
res = mod.fit()
print(res.summary())
df['xcube'] = df['x']**3  

mod2= smf.ols('y ~ x + xsqr', data=df)
res2 = mod2.fit()
print(res2.summary())

mod3= smf.ols('y ~ x + xsqr + xcube', data=df)
res3 = mod3.fit()
print(res2.summary())

dfoutput = summary_col([res,res2,res3],stars=True)
print(dfoutput)
-------------------------------------------------------------
Which has the following output:

=====================================
            y I       y II    y III  
-------------------------------------
Intercept 0.9909   -0.6576   -0.2904 
          (0.9083) (1.0816)  (1.3643)
x         0.3732*  1.7776*** 1.0700  
          (0.2068) (0.6236)  (1.6736)
xcube                        -0.0184 
                             (0.0402)
xsqr               -0.1845** 0.0409  
                   (0.0781)  (0.4995)
=====================================
Standard errors in parentheses.
* p<.1, ** p<.05, ***p<.01
```
