CSV cleaning for research

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as pdr
import datetime



econ_data = 'Money_Stringency_Vax_data.csv'
df1 = pd.read_csv(econ_data) #, dtype = {'Date': str})
#print(df1)
```

Checking for any columns to drop. i.e. Unnamed:21, Unnamed:22

```python
df1.describe()
```

```python
df1.drop('Unnamed: 21',axis=1,inplace=True)
df1.drop('Unnamed: 22',axis=1,inplace=True)
df1.drop('Unnamed: 60',axis=1,inplace=True)
```

To drop the weird header names I am going to drop them and carry up the first row

```python
new_header = df1.iloc[0] #grab first row for the header
df1 = df1[1:] # take data less the header row
df1.columns = new_header

list(df1.columns) # checking 
```

```python
df1.drop('date_vac',axis=1,inplace=True) # dropping vac date column becuase it is the same as the first date column
```

Pulling data for Consumer Confidence Index by UMICH

```python
# Didnt know when the cci started, so I chose an arbitrary start
start = datetime.datetime(1900,1,1)  
end = datetime.datetime.now()

df2 = pdr.DataReader('UMCSENT', start = start, end = end, data_source = 'fred')
df2 = df2.stack().reset_index()
print(df2) # checking
# this saves the file
#df2 = df2.to_csv('UMCSENT_historical1.csv', encoding='utf-8', index=False)
```
